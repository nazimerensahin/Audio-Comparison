```python
# Librosa sisteminin aktarlmas
import matplotlib.pyplot as plt # Grafik ve grselletirme ilemleri iin Matplotlib ktphanesini ykler.
import numpy as np # Bilimsel hesaplama ve dizi ilemleri iin NumPy ktphanesini ykler.
import librosa # Ses ileme ve analiz iin Librosa ktphanesini ykler.
import librosa.display # Librosann ses grafikleri izme fonksiyonlarn ykler.
import pyloudnorm as pyln  # LUFS analizini yapar.
import IPython.display as ipd # Ses dosyalarn Jupyter'da almak iin gerekli IPython fonksiyonlarn ykler.
from IPython.display import Audio, display  # Ses dosyalarn Jupyter hcresinde oynatmaya olanak tanr.

# Ses dosyalarnn yollarn bir listeye ekleme
audio_files = [
    "/Users/nazimerensahin/Desktop/fldr/teletro.wav",  
    "/Users/nazimerensahin/Desktop/fldr/22.wav",      
    "/Users/nazimerensahin/Desktop/fldr/33.wav",
    "/Users/nazimerensahin/Desktop/fldr/44.wav",
    "/Users/nazimerensahin/Desktop/fldr/55.wav",
    "/Users/nazimerensahin/Desktop/fldr/66.wav",
    "/Users/nazimerensahin/Desktop/fldr/77.wav",
    "/Users/nazimerensahin/Desktop/fldr/88.wav",
    "/Users/nazimerensahin/Desktop/fldr/99.wav",
    "/Users/nazimerensahin/Desktop/fldr/1010.wav",
    "/Users/nazimerensahin/Desktop/fldr/1111.wav",
]

# Her dosya iin aklayc isimler
audio_labels = [
    "Teletronix Input",
    "Processed Sound 1",
    "Processed Sound 2",
    "Processed Sound 3",
    "Processed Sound 4",
    "Processed 2 5",
    "Processed Sound 6",
    "Processed Sound 7",
    "Processed Sound 8",
    "Processed Sound 9",
    "Processed Sound 10",
]

# THD hesaplama fonksiyonu
def thd_analysis(signal, sr):
    ft = np.fft.fft(signal)  # Fourier Transform
    magnitude = np.abs(ft)  # Byklk spektrumu
    fundamental_freq = np.argmax(magnitude[:len(magnitude)//2])  # Temel frekans
    
    # Harmonik frekanslar ve toplam harmonik bozulma (THD)
    harmonics = [np.argmax(magnitude[n * fundamental_freq - 5 : n * fundamental_freq + 5])
                 for n in range(2, 6)]  # 2., 3., 4., 5. harmonikler
    thd = sum(magnitude[harmonics]) / magnitude[fundamental_freq]
    return thd

# Dosyalar ykle ve analiz etme
sound1, sr1 = librosa.load(audio_files[0], sr=None)
thd_values = [thd_analysis(sound1, sr1)]  # lk dosyann THD deeri

# Dier ses dosyalarn srayla analiz et
for file in audio_files[1:]:
    sound, sr = librosa.load(file, sr=None)
    thd_values.append(thd_analysis(sound, sr))

# THD sonularn grselletirme
plt.figure(figsize=(12, 6))
plt.bar(audio_labels, thd_values, color='skyblue')  # audio_labels kullanlyor
plt.ylabel("THD Deeri")
plt.title("Total Harmonic Distortion (THD) Karlatrmas")
plt.xticks(rotation=45)  # Etiketleri dndr
plt.ylim(0, max(thd_values) * 1.2)  # Grafik snrlarn belirle
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
