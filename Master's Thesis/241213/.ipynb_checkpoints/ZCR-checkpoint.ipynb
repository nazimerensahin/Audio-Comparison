```python
# ZCR, bir ses sinyalinin sfrdan gei saysn (pozitiften negatife veya negatiften pozitife geiler) hesaplar. 
# zellikle bir sesin perkusif (vurulu) mi, srekli mi, veya grlt tabanl m, tonal mi olduunu anlamak iin kullanlr.
# Y ekseni, sinyalin sfr geilerinin ereve iindeki toplam rnek saysna orandr ve genelde 0 ile 1 arasnda bir lek iinde hesaplanr.

# nerilen dosya: akustik davul

# Gerekli ktphanelerin yklenmesi
import matplotlib.pyplot as plt
import librosa
import librosa.display

# Ses dosyalarnn yollarn ve aklayc isimlerini tanmlama
audio_files = [
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/LA2A Teletronix_12.wav", # Teletronix LA-2A
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/LA2A Urei_11.wav", # Urei LA-2A    
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/TheLongWait.drums_uad_gray.wav", # UAD LA-2A Gray
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/TheLongWait.drums uad silver render_03.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
    "/Users/nazimerensahin/Desktop/d100hz.wav",
]

# Ses dosyalarnn isimlendirilmesi
audio_labels = [
    "Teletronix LA-2A",
    "Urei LA-2A",
    "UAD LA-2A Gray",
    "UAD LA-2A Silver",
    "UAD LA-2A Tube",
    "Waves CLA-2A",
    "IKM T-RackS White 2A",
    "NI Softube VC 2A",
    "Slate Digital FG-2A",
    "BRA VLA-2A",
    "Overloud Gem Comp LA",
    "Analog Obsession LALA",
]

# Teletronix LA-2A iin referans dalgalar
teletronix_wave, teletronix_sr = librosa.load(audio_files[0], sr=None)
teletronix_zcr = librosa.feature.zero_crossing_rate(teletronix_wave)[0]
teletronix_time = librosa.times_like(teletronix_zcr, sr=teletronix_sr)

# Urei LA-2A ve UAD LA-2A Silver karlatrmas
urei_wave, urei_sr = librosa.load(audio_files[1], sr=None)
urei_zcr = librosa.feature.zero_crossing_rate(urei_wave)[0]
urei_time = librosa.times_like(urei_zcr, sr=urei_sr)

silver_wave, silver_sr = librosa.load(audio_files[3], sr=None)
silver_zcr = librosa.feature.zero_crossing_rate(silver_wave)[0]
silver_time = librosa.times_like(silver_zcr, sr=silver_sr)

# Grafik ayarlar
plt.figure(figsize=(16, 12))
plot_idx = 1

# 1. Grafik: Urei LA-2A ve UAD LA-2A Silver
plt.subplot(5, 2, plot_idx)
plt.plot(urei_time, urei_zcr, color="blue", label=audio_labels[1])
plt.plot(silver_time, silver_zcr, color="red", label=audio_labels[3])
plt.title(f"{audio_labels[1]} ve {audio_labels[3]}")
plt.xlabel("Zaman (saniye)")
plt.ylabel("Zero-Crossing Oran")
plt.legend()
plt.grid()
plot_idx += 1

# kinci ve Sonras: Teletronix LA-2A ile dier pluginler
for idx, (file_path, label) in enumerate(zip(audio_files[2:], audio_labels[2:]), start=2):
    if label in [audio_labels[1], audio_labels[3]]:  # Urei ve Silver zaten karlatrld
        continue

    # Dier ses dosyasn ykleme
    other_wave, other_sr = librosa.load(file_path, sr=None)
    other_zcr = librosa.feature.zero_crossing_rate(other_wave)[0]
    other_time = librosa.times_like(other_zcr, sr=other_sr)

    # Karlatrma grafii
    plt.subplot(5, 2, plot_idx)
    plt.plot(teletronix_time, teletronix_zcr, color="blue", label=audio_labels[0])
    plt.plot(other_time, other_zcr, color="red", label=label)
    plt.title(f"{audio_labels[0]} ve {label}")
    plt.xlabel("Zaman (saniye)")
    plt.ylabel("Zero-Crossing Oran")
    plt.legend()
    plt.grid()
    plot_idx += 1

# Genel dzenleme ve gsterim
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.suptitle("Zero-Crossing Oran", fontsize=16)
plt.show()
```
