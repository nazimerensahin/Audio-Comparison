```python
# Referans bir ses sinyali ile dier ses sinyalleri arasndaki faz farklarn karlatrmak ve grafiksel olarak sunmaktr.
# ki ses sinyali arasndaki faz farkn radian cinsinden renk haritasyla gsteriyor.
# Krmz (pozitif rad): ki sinyal arasnda faz fark var ve bir sinyal dierinin gerisinde.
# Mavi (negatif rad): kinci sinyal nde.
# Beyaz veya ntr renk: Faz fark az veya yok.

# nerilen dosya: white noise

# Librosa sisteminin aktarlmas
import matplotlib.pyplot as plt # Grafik ve grselletirme ilemleri iin Matplotlib ktphanesini ykler.
import numpy as np # Bilimsel hesaplama ve dizi ilemleri iin NumPy ktphanesini ykler.
import librosa # Ses ileme ve analiz iin Librosa ktphanesini ykler.
import librosa.display # Librosann ses grafikleri izme fonksiyonlarn ykler.
import pyloudnorm as pyln  # LUFS analizini yapar.
import IPython.display as ipd # Ses dosyalarn Jupyter'da almak iin gerekli IPython fonksiyonlarn ykler.
from IPython.display import Audio, display  # Ses dosyalarn Jupyter hcresinde oynatmaya olanak tanr.
from scipy import signal # Python'daki SciPy ktphanesinin signal modln ie aktarmaya yarar.

# Ses dosyalarnn yollar
audio_files = [
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Teletronix_05.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Urei_04.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad gray white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad silver white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Teletronix_05.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Urei_04.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad gray white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad silver white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Teletronix_05.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Urei_04.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad gray white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad silver white noise.wav",
]

# Dosya etiketleri
audio_labels = [
    "Teletronix LA-2A",
    "Urei LA-2A",
    "UAD LA-2A Gray",
    "UAD LA-2A Silver",
    "UAD LA-2A Tube",
    "Waves CLA-2A",
    "IKM T-RackS White 2A",
    "NI Softube VC 2A",
    "Slate Digital FG-2A",
    "BRA VLA-2A",
    "Overloud Gem Comp LA",
    "Analog Obsession LALA",
]

# Karlatrma kurallar
main_reference = 0  # Teletronix LA-2A
urei_index = 1  # Urei LA-2A
silver_index = 3  # UAD LA-2A Silver

# Faz fark grselletirme fonksiyonu
def plot_phase_difference(y_ref, sr_ref, y_comp, sr_comp, title, ax):
    # STFT hesaplama
    S_ref = librosa.stft(y_ref)
    S_comp = librosa.stft(y_comp)

    # Faz fark hesaplama
    phase_ref = np.angle(S_ref)
    phase_comp = np.angle(S_comp)
    phase_diff = phase_ref - phase_comp

    # Faz farkn grselletirme
    img = librosa.display.specshow(
        phase_diff, sr=sr_ref, x_axis="time", y_axis="log", cmap="coolwarm", ax=ax
    )
    ax.set_title(title)
    plt.colorbar(img, ax=ax, format="%+2.0f rad", label="Faz Fark")

# Grafik dzeni iin alt eksen ayar
n_comparisons = len(audio_files) - 2  # Urei ve Silver karlatrmalarn hesaba kat
n_cols = 2
n_rows = (n_comparisons + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 2))
axes = axes.flatten()

plot_idx = 0

# Urei LA-2A ile sadece UAD Silver karlatrmas
y_urei, sr_urei = librosa.load(audio_files[urei_index], sr=None)
y_silver, sr_silver = librosa.load(audio_files[silver_index], sr=None)

plot_phase_difference(
    y_urei,
    sr_urei,
    y_silver,
    sr_silver,
    f"{audio_labels[urei_index]} ve {audio_labels[silver_index]} Faz Fark",
    axes[plot_idx],
)
plot_idx += 1

# Teletronix LA-2A ile dier karlatrmalar
y_main_ref, sr_main_ref = librosa.load(audio_files[main_reference], sr=None)

for idx, (file, label) in enumerate(zip(audio_files, audio_labels)):
    if idx in {main_reference, urei_index, silver_index}:
        continue  # Urei ve Silver, farkl kurallara gre karlatrld

    # Dosyay ykle
    y_comp, sr_comp = librosa.load(file, sr=None)

    # Grafik izimi
    plot_phase_difference(
        y_main_ref,
        sr_main_ref,
        y_comp,
        sr_comp,
        f"{audio_labels[main_reference]} ve {label} Faz Fark",
        axes[plot_idx],
    )
    plot_idx += 1

# Bo kalan eksenleri silme
for idx in range(plot_idx, len(axes)):
    fig.delaxes(axes[idx])

# Genel dzenleme ve gsterim
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.suptitle("Faz Fark", fontsize=16)
plt.show()
```
