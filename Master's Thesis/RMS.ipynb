```python
# Farkl ses sinyallerinin ortalama genlik deerlerini grafik erisi ile grselletirir.
# Ses dosyalar genellikle normalletirilir, bu da genlik deerlerinin aralnn -1 ile 1 arasnda olaca anlamna gelir.
# Y ekseninde olan snrlar ses dosyasnn bozulmadan (clipping olmadan) ilenmesini salar.

# White noise, frekans tepkisi ve dinamik sktrma tepkisini ayn anda inceleyebilmeyi salar. 
# zellikle LA-2A'nn non-lineer davran ve harmonik distorsiyon eklemelerini analiz ederken en doru sinyal tipidir.

# Librosa sisteminin aktarlmas
import matplotlib.pyplot as plt # Grafik ve grselletirme ilemleri iin Matplotlib ktphanesini ykler.
import numpy as np # Bilimsel hesaplama ve dizi ilemleri iin NumPy ktphanesini ykler.
import librosa # Ses ileme ve analiz iin Librosa ktphanesini ykler.
import librosa.display # Librosann ses grafikleri izme fonksiyonlarn ykler.
import pyloudnorm as pyln  # LUFS analizini yapar.
import IPython.display as ipd # Ses dosyalarn Jupyter'da almak iin gerekli IPython fonksiyonlarn ykler.
from IPython.display import Audio, display  # Ses dosyalarn Jupyter hcresinde oynatmaya olanak tanr.
from scipy import signal # Python'daki SciPy ktphanesinin signal modln ie aktarmaya yarar.

# Ses dosyalarnn yollarn bir listeye ekleme
audio_files = [
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Teletronix_05.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Urei_04.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad gray white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad silver white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Teletronix_05.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Urei_04.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad gray white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad silver white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Teletronix_05.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/white.noise.LA2A Urei_04.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad gray white noise.wav",
    "/Users/nazimerensahin/Desktop/la_2a_processed/PLUGINS/Audio Files/uad silver white noise.wav",
]

# Her dosya iin aklayc isimler ekleme
audio_labels = [
    "Teletronix LA-2A",
    "Urei LA-2A",
    "UAD LA-2A Gray",
    "UAD LA-2A Silver",
    "UAD LA-2A Tube",
    "Waves CLA-2A",
    "IKM T-RackS White 2A",
    "NI Softube VC 2A",
    "Slate Digital FG-2A",
    "BRA VLA-2A",
    "Overloud Gem Comp LA",
    "Analog Obsession LALA",
]

# Teletronix LA-2A referans olarak ykleme
ref_file = audio_files[0]
ref_label = audio_labels[0]
ref_sound, ref_sr = librosa.load(ref_file, sr=None)

# RMS hesaplama fonksiyonu
def compute_rms(sound, sr, frame_length=2048, hop_length=512):
    rms = librosa.feature.rms(y=sound, frame_length=frame_length, hop_length=hop_length)[0]
    time = librosa.times_like(rms, sr=sr, hop_length=hop_length)
    return rms, time

# Referans RMS hesaplamalar
ref_rms, ref_time = compute_rms(ref_sound, ref_sr)

# Urei LA-2A dosyasn ykleme ve RMS hesaplama
urei_file = audio_files[1]
urei_label = audio_labels[1]
urei_sound, urei_sr = librosa.load(urei_file, sr=None)
urei_rms, urei_time = compute_rms(urei_sound, urei_sr)

# UAD LA-2A Silver dosyasn ykleme ve RMS hesaplama
silver_file = audio_files[3]
silver_label = audio_labels[3]
silver_sound, silver_sr = librosa.load(silver_file, sr=None)
silver_rms, silver_time = compute_rms(silver_sound, silver_sr)

# Grafik dzeni: 10 grafik olacak
n_cols = 2
n_rows = 5

# Grafik figr oluturma
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 3))

# oklu eksenler iin 2D dzeni 1D olarak ileme
axes = axes.flatten()

# Urei ve UAD Silver karlatrmas
ax = axes[0]
librosa.display.waveshow(urei_sound, sr=urei_sr, alpha=0.4, color='red', ax=ax, label=f"{urei_label} Waveform")
ax.plot(urei_time, urei_rms, color="blue", linewidth=1, label=f"{urei_label} RMS")
librosa.display.waveshow(silver_sound, sr=silver_sr, alpha=0.4, color='green', ax=ax, label=f"{silver_label} Waveform")
ax.plot(silver_time, silver_rms, color="yellow", linewidth=1, label=f"{silver_label} RMS")
ax.set_title(f"{urei_label} ve {silver_label}")
ax.set_xlabel("Time (seconds)")
ax.set_ylabel("Amplitude")
ax.set_ylim((-0.1, 0.5))
ax.legend(fontsize=8, loc="upper right", frameon=True, framealpha=0.9, borderpad=1)

# Dier pluginleri Teletronix ile karlatrma
valid_idx = 1  # Grafik iin kullanlacak eksen indeksi (ilk grafik dolu)
for file, label in zip(audio_files[2:], audio_labels[2:]):
    # Urei ve Silver' atla
    if label in [urei_label, silver_label]:
        continue
    
    sound, sr = librosa.load(file, sr=None)
    rms, time = compute_rms(sound, sr)

    ax = axes[valid_idx]
    librosa.display.waveshow(ref_sound, sr=ref_sr, alpha=0.4, color='red', ax=ax, label=f"{ref_label} Waveform")
    ax.plot(ref_time, ref_rms, color="blue", linewidth=1, label=f"{ref_label} RMS")
    librosa.display.waveshow(sound, sr=sr, alpha=0.4, color='green', ax=ax, label=f"{label} Waveform")
    ax.plot(time, rms, color="yellow", linewidth=1, label=f"{label} RMS")
    ax.set_title(f"{label} ve {ref_label}")
    ax.set_xlabel("Time (seconds)")
    ax.set_ylabel("Amplitude")
    ax.set_ylim((-0.1, 0.5))
    ax.legend(fontsize=8, loc="upper right", frameon=True, framealpha=0.9, borderpad=1)
    
    valid_idx += 1  # Sonraki eksene ge

# Bo kalan alt eksenleri gizleme
for idx in range(valid_idx, len(axes)):
    fig.delaxes(axes[idx])

# Genel dzenleme ve gsterim
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.suptitle("Root Mean Square Karlatrmas", fontsize=16)
plt.show()
```
