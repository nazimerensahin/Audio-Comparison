```python
# Librosa sisteminin aktarlmas

import matplotlib.pyplot as plt # Grafik ve grselletirme ilemleri iin Matplotlib ktphanesini ykler.
import numpy as np # Bilimsel hesaplama ve dizi ilemleri iin NumPy ktphanesini ykler.
import librosa # Ses ileme ve analiz iin Librosa ktphanesini ykler.
import librosa.display # Librosann ses grafikleri izme fonksiyonlarn ykler.
import pyloudnorm as pyln  # LUFS analizini yapar.
import IPython.display as ipd # Ses dosyalarn Jupyter'da almak iin gerekli IPython fonksiyonlarn ykler.
from IPython.display import Audio, display  # Ses dosyalarn Jupyter hcresinde oynatmaya olanak tanr.

# Ses dosyalarnn yollarn bir listeye ekleme
audio_files = [
    "/Users/nazimerensahin/Desktop/fldr/teletro.wav",  
    "/Users/nazimerensahin/Desktop/fldr/22.wav",      
    "/Users/nazimerensahin/Desktop/fldr/33.wav",
    "/Users/nazimerensahin/Desktop/fldr/44.wav",
    "/Users/nazimerensahin/Desktop/fldr/55.wav",
    "/Users/nazimerensahin/Desktop/fldr/66.wav",
    "/Users/nazimerensahin/Desktop/fldr/77.wav",
    "/Users/nazimerensahin/Desktop/fldr/88.wav",
    "/Users/nazimerensahin/Desktop/fldr/99.wav",
    "/Users/nazimerensahin/Desktop/fldr/1010.wav",
    "/Users/nazimerensahin/Desktop/fldr/1111.wav",
]

# Her dosya iin aklayc isimler ekleme
audio_labels = [
    "Teletronix Input",
    "Processed Sound 1",
    "Processed Sound 2",
    "Processed Sound 3",
    "Processed Sound 4",
    "Processed Sound 5",
    "Processed Sound 6",
    "Processed Sound 7",
    "Processed Sound 8",
    "Processed Sound 9",
    "Processed Sound 10",
]

# Dinamik Aralk (DR) Hesaplama Fonksiyonu
def calculate_dynamic_range(signal):
    rms = librosa.feature.rms(y=signal)
    dr = np.max(rms) / np.min(rms)
    return 20 * np.log10(dr)  # dB cinsinden dinamik aralk

# Dinamik Aralk ve LUFS Deerlerini Hesaplama
sound1, sr1 = librosa.load(audio_files[0], sr=None)  # Sound1'i ykle
meter1 = pyln.Meter(rate=sr1)  # Sound1 iin bir loudness lm arac olutur

dr_values = []  # Dinamik aralk deerleri
lufs_values = []  # LUFS deerleri
labels = []  # Ses dosyas isimleri

for i, file in enumerate(audio_files):
    sound, sr = librosa.load(file, sr=None)  # Ses dosyasn ykle
    meter = pyln.Meter(rate=sr)  # Ses dosyasna uygun meter olutur
    
    # Dinamik aralk ve LUFS hesaplama
    dr = calculate_dynamic_range(sound)
    lufs = meter.integrated_loudness(sound)
    
    # Deerleri saklama
    dr_values.append(dr)
    lufs_values.append(lufs)
    labels.append(f"Sound{i + 1}")

# Grafik izimi
x = np.arange(len(labels))  # ubuklarn x konumlar
width = 0.35  # ubuk genilii

fig, ax = plt.subplots(figsize=(14, 5))  # Grafik boyutlar

# Dinamik Aralk ubuklar
bar1 = ax.bar(x - width / 2, dr_values, width, label='Dinamik Aralk (DR)', color='skyblue')

# LUFS ubuklar
bar2 = ax.bar(x + width / 2, lufs_values, width, label='LUFS', color='salmon')

# Grafik etiketleri ve balk
ax.set_ylabel('Deer')
ax.set_title('Ses Dosyalarnn Dinamik Aralk ve LUFS Karlatrmas')
ax.set_xticks(x)
ax.set_xticklabels(labels, rotation=45)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.18), ncol=2) # st ksmda bulunan legend "lejant" ksmnn konumunu belirler.

# ubuklarn zerine deerleri yazdrma
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 0),  # ubuun stne yazdr
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bar1)
add_labels(bar2)

plt.tight_layout()
plt.show()
```
